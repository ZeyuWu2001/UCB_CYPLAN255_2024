# Part 3: Using Pandas to Clean Data

```python
import pandas as pd
```

```python
# Cleaning National Park Service boundary dataset to isolate for Californian national parks only
park_boundaries = pd.read_csv('HW4_Data/NPS_Land_Resources_Division_Boundary.csv')

filtered_park_boundaries = park_boundaries[(park_boundaries['STATE'] == 'CA') & (park_boundaries['UNIT_TYPE'] == 'National Park')]
filtered_park_boundaries.head(10)
```

```python
# Cleaning National Park Service monthly visitor dataset for Yosemite National Park
# Filtering the data to the peak months of May to October, inclusive
# Creating a dataset for post-COVID visitor data

yosemite_visitors = pd.read_csv('HW4_Data/Yosemite_Visitation_by_Month.csv')
yosemite_visitors_peak = yosemite_visitors[['Year','MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT']]

yosemite_visitors_peak_postCovid = yosemite_visitors_peak[yosemite_visitors_peak['Year'] > 2020]

months = ['MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT']
for month in months:
    yosemite_visitors_peak_postCovid[month] = yosemite_visitors_peak_postCovid[month].str.replace(',', '').astype(int)
```

```python
# Filtering National Park Boundary shapefile dataset to show just Yosemite to pull into Replicahq.com for OD data

import geopandas as gpd
all_parks = gpd.read_file('HW4_Data/NPS_Shapefiles/nps_boundary.shp')

# Assuming the shapefile has an attribute 'Park_Name' containing park names
# Filter the dataframe to select only the rows for Yosemite National Park
yosemite_shapefile = all_parks[all_parks['PARKNAME'] == 'Yosemite']

# Save the Yosemite National Park shapefile to a new file
yosemite_shapefile.to_file('HW4_Data/NPS_Shapefiles/yosemite_shapefile.shp')
```

# Part 4: Using group-by

```python
# Part 4: Using group-by operations to produce pivot table for Yosemite National Park monthly visitor data by month, post-COVID

yosemite_visitors_peak_postCovid.groupby(['Year']).sum()
```

```python
#Seeing the maximum visitor count from all months in May
yosemite_visitors_peak_postCovid['MAY'].max()
```

# Part 5: Using Matplotlib or Seaborn

```python
# Part 5: Using Matplotlib or Seaborn to produce data visualizations
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline

months = ['MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT']
bar_width = 0.15
n_years = len(yosemite_visitors_peak_postCovid['Year'].unique())

for i, (year, data) in enumerate(yosemite_visitors_peak_postCovid.groupby('Year')):
    x = np.arange(len(months)) + i * (bar_width)  # Adjusting x positions for each year
    plt.bar(x, data[months].values.ravel(), width=bar_width, label=year)
   
plt.xlabel('Month')
plt.ylabel('Monthly Visitors')
plt.title('Yosemite Peak Season Monthly Visitors by Year, 2021 to 2023')
plt.legend()
plt.xticks(np.arange(len(months)) + bar_width, months)
plt.show()
```

# Part 6: Data Interpretation

By looking at the Yosemite peak season (May to October) post-COVID visitor data, the months with the highest overall visitors are July and August, and visitor numbers tapering off around these months. We can see a gradual increase in visitors as we move further away from 2020 (COVID debut year), perhaps with COVID restrictions easing and greater vaccination rates. Summer visitor data drops are seen in the months following July 2021 as well as a singular dip in July 2022; these declines could be associated with the summer COVID surges experienced nationwide.
